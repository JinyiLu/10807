```python
class Box:
    def __init__(self, left, top, right, bottom, label, conf):
        self.left = left
        self.top = top
        self.right = right
        self.bottom = bottom
        self.label = label
        self.conf = conf

def box_avg(boxes, conf_th=0.5):
    avg_w = []
    avg_h = []
    avg_area = []
    cnt = 0
    for b in boxes:
        if b.conf > conf_th:
            avg_w.append(b.right - b.left)
            avg_h.append(b.bottom - b.top)
            avg_area.append((b.right - b.left) * (b.bottom - b.top))
            cnt += 1
    print 'avg w %f avg h %f avg area %f cnt %d' % (np.mean(avg_w), np.mean(avg_h), np.mean(avg_area), len(avg_w))
#     print 'std w %f std h %f std area %f' % (np.std(avg_w), np.std(avg_h), np.std(avg_area))
    return avg_w, avg_h, avg_area, cnt

def cal_avg_rectangle(mask, result_path, truth=True):
    valid = []
    f = open(mask, 'r')
    boxes = {}
    for l in f:
        img = l.strip().replace('.png', '.txt')
        valid.append(img)
        fi = open(result_path+img.strip(), 'r')
        for li in fi:
            sp = li.split()
            conf = 0.0
            if truth:
                conf = 1.0
            else:
                conf = float(sp[-1])
            b = Box(float(sp[4]), float(sp[5]), float(sp[6]), float(sp[7]), sp[0], conf)
            if sp[0] not in boxes:
                boxes[sp[0]] = [b]
            else:
                boxes[sp[0]].append(b)
        
    print len(valid)
    
    return boxes, valid
    
```

```python
boxes, valid = cal_avg_rectangle('../../data/valid_30_mask.txt', '../../data/object/label_2/')

print len(boxes['Car']), len(boxes['Car']) / float(len(valid))
print len(boxes['Pedestrian']), len(boxes['Pedestrian']) / float(len(valid))
print len(boxes['Cyclist']), len(boxes['Cyclist']) / float(len(valid))

box_avg(boxes['Car'], 0.5)
box_avg(boxes['Pedestrian'], 0.5)
print '**'
# print box_avg(boxes['Cyclist'], 0.5)
```
